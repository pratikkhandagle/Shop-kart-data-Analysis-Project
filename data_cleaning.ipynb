```python
plt.figure(figsize=(15, 5))

cols = ["Age","Profit","Selling_Price",'Revenue']

# Ensure 'cols' is a list of your column names
for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)  # 1 row, 3 columns, 'i' is the current plot position
    sns.boxplot(y=data[col], color='skyblue', width=0.3)
    plt.ylabel('')
    plt.title(f'Boxplot of {col}') # Optional: adds a title to each subplot

plt.tight_layout() # Helps prevent overlapping
plt.show()
```
